```python
from transformers import AutoModelForCausalLM, AutoTokenizer;


# 加载预训练模型和分词器
model = AutoModelForCausalLM.from_pretrained("microsoft/deberta-base", # 
                                             device_map="cuda", # 如果有GPU则使用GPU    
                                                torch_dtype="auto", # 自动选择最佳数据类型
                                                trust_remote_code=True);

# 加载语言模型的分词器
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-xsmall", trust_remote_code=True);

# 对句子进行分词
inputs = tokenizer("Hello, my dog is cute", return_tensors="pt");

# 使用模型进行前向传播，获取输出
outputs = model(**inputs);

print(outputs);

# 获取最后一个隐藏状态
print(outputs.last_hidden_state);
```

```python
from sentence_transformer import SentenceTransformer  ;

# 加载 SentencePiece 模型
model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2");

# 对文本进行分词
tokens = model.encode("Hello, my dog is cute", out_type=str);

print(tokens);
```
