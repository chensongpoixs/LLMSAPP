```python
import numpy as np

def embedding_layer(text, embedding_dim):
  """
  计算一个文本的嵌入向量。

  Args:
    text: 输入文本 (字符串)。
    embedding_dim: 嵌入向量的维度 (例如，300)。

  Returns:
    一个包含文本嵌入向量的 NumPy 数组。
  """
  # 1. 预处理 (简单处理，可以根据需要进行更复杂的预处理)
  words = text.split()  # 简单分割成单词列表
  word_vectors = []
  for word in words:
    word_vectors.append(np.random.rand(300))  # 随机生成一个300维的词向量

  print(f"单词列表: {words}");
  print(f"词向量列表: {word_vectors}");

  # 2. 嵌入向量 ==> 计算文本的嵌入向量 (这里简单地取平均作为示例)
  embeddings = np.dot(word_vectors, np.random.rand(300))  # 使用矩阵乘法计算
  return embeddings


# 示例用法
text = "今天天气真好。"
embedding_dim = 300;
embedding = embedding_layer(text, embedding_dim);

print(f"文本: {text}");
print(f"嵌入向量: {embedding}");
 

# **代码解释:**

# *   `embedding_layer(text, embedding_dim)` 函数接收文本和嵌入维度作为输入。
# *   `words = text.split()`: 将文本分割成单词列表。
# *   `word_vectors = []`:  初始化一个空列表，用于存储每个单词的嵌入向量。
# *   `for word in words:`: 遍历单词列表。
# *   `word_vectors.append(np.random.rand(300))`:  生成一个300维的随机词向量，并添加到 `word_vectors` 列表中。  (这个过程 可以根据需要进行更复杂的词向量生成
```
