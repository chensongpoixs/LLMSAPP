```python
import Tokenizer;

# Initialize the tokenizer with the specified text file
tokenizer = Tokenizer.SimpleTokenizer("../LLM/data/华盖集_cleaned.txt");

# print(tokenizer.tokenize("人生自古谁无死，留取丹心照汗青。"));
ids = tokenizer.encode("人生自古谁无死，留取丹心照汗青。");

# Print the token IDs
print(ids);

# Decode the token IDs back to text
text = tokenizer.decode(ids);
print(text);
```
